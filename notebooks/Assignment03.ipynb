```python
# import python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as md
%matplotlib inline
```

```python
# load wave data at Hilo, HI, see its location here: https://cdip.ucsd.edu/m/products/?stn=188p1&param=waveHs
ds_hilo = pd.read_csv('https://erddap.cdip.ucsd.edu/erddap/tabledap/wave_agg.csv?station_id%2Ctime%2CwaveHs%2CwaveTp%2CwaveTa%2CwaveDp%2CmetaStationName%2Clatitude%2Clongitude&station_id=%22188%22&time%3E2025-09-04T00%3A00%3A00Z&waveFlagPrimary=1')
ds_hilo
```

```python
# load wave data at Oahu, HI, see its location here: https://cdip.ucsd.edu/m/products/?stn=098p1&param=waveHs
ds_oahu = pd.read_csv('https://erddap.cdip.ucsd.edu/erddap/tabledap/wave_agg.csv?station_id%2Ctime%2CwaveHs%2CwaveTp%2CwaveTa%2CwaveDp%2CmetaStationName%2Clatitude%2Clongitude&station_id=%22098%22&time%3E2025-09-04T00%3A00%3A00Z&waveFlagPrimary=1')
ds_oahu
```

```python
# plot significant wave height
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,3))
ax.plot(pd.to_datetime(ds_hilo["time"][1::]), ds_hilo["waveHs"][1::].to_numpy(dtype=np.float64), marker='.', label = 'HILO, HAWAII, HI ')
# ax.plot(pd.to_datetime(ds_oahu["time"][1::]), ds_oahu["waveHs"][1::].to_numpy(dtype=np.float64), marker='.', label = 'MOKAPU POINT, HI ')
ax.grid(True)
ax.legend()
ax.set_xlabel('time, UTC')
ax.set_ylabel('significant wave height [m]')
# plot date-hours-minutes-seconds
xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)
plt.tight_layout()
plt.xticks( rotation=25 )
plt.minorticks_on()
plt.show()
```

```python
# plot peak wave period
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,3))
ax.plot(pd.to_datetime(ds_hilo["time"][1::]), ds_hilo["waveTp"][1::].to_numpy(dtype=np.float64), marker='.', label = 'HILO, HAWAII, HI ')
# ax.plot(pd.to_datetime(ds_oahu["time"][1::]), ds_oahu["waveTp"][1::].to_numpy(dtype=np.float64), marker='.', label = 'MOKAPU POINT, HI ')
ax.grid(True)
ax.legend()
ax.set_xlabel('time, UTC')
ax.set_ylabel('peak wave period [s]')
# plot date-hours-minutes-seconds
xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)
plt.tight_layout()
plt.xticks( rotation=25 )
plt.minorticks_on()
plt.show()
```

```python
# plot peak wave direction: 0/360 degree - due east; 90 degree - due north; 180 degree - due west; 270 degree - due south
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8,3))
ax.plot(pd.to_datetime(ds_hilo["time"][1::]), ds_hilo["waveDp"][1::].to_numpy(dtype=np.float64), marker='.', label = 'HILO, HAWAII, HI ')
# ax.plot(pd.to_datetime(ds_oahu["time"][1::]), ds_oahu["waveDp"][1::].to_numpy(dtype=np.float64), marker='.', label = 'MOKAPU POINT, HI ')
ax.grid(True)
ax.legend()
ax.set_xlabel('time, UTC')
ax.set_ylabel('peak wave direction [degree]')
# plot date-hours-minutes-seconds
xfmt = md.DateFormatter('%Y-%m-%d %H:%M:%S')
ax.xaxis.set_major_formatter(xfmt)
plt.tight_layout()
plt.xticks( rotation=25 )
plt.minorticks_on()
plt.show()
```
