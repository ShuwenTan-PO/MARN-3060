## 1. Basic sediment properties

### 1). Notation and basic properties

- Sediment mass and volume: $m_s$, $V_s$
- Sediment density: $\rho_s = \frac{m_s}{V_s}$
- Specific weight (=weight per unit volume): $\gamma_s=\frac{W_s}{V_s}=\frac{\rho_s gV_s}{V_s}=\rho_s g$
- Specific gravity: $s=\frac{\rho_s g}{\rho_0 g} = \frac{\gamma_s}{\gamma_0}$, e.g. Quartz: s = 2.65 times more dense than water at 24$^{\circ}$C.
- Porosity $p$: the ratio of pore space (voids) to the whole sediment volume. Natural sands have porosities in the range of 0.25 to 0.50; a frequently applied figure is 0.40 (or 40 %); for maximum possible sphere packing: $p$=0.64; for M&M candies, $p$=0.68.
- Sediment concentration $c$ can be defined in two ways: mass concentration
and volume concentration.
    - The mass concentration is the mass of the solid
particles per volume ($c$ in kg/m$^3$ or equivalently g/L) and is often used when
measuring sediment concentrations.
    - Volume concentration is defined as the ratio
of the volume of solid particles to the whole volume ($c$ in m$^3$/m$^3$ or in %).
Sediment in a sediment bed has a volume concentration of $n$ = 1 âˆ’ $p$. For sediment
in suspension, the volume concentration $c$ indicates the volume of sediment
per volume of the mixture. If the sediment in such a mixture settles to
the bed, 1 m$^3$ of solid particles will occupy 1/(1 âˆ’ $p$) ($p$: porosity) m$^3$ at the bed.
Volume concentrations are obtained from mass concentrations by multiplication
by 1/$\rho_s$;

### 3). Sediment distribution

#### a. Particle size distribution (PSD or GSD)

Sediment grain sizes distributions are typically approximated as lognormal
with mean $\mu$ and standard deviation $\sigma$.

$d_N$ = diameter for which $N\%$ of grains by mass are smaller.

e.g. $d_{50}$ = median grain size; 50% of grains by mass are smaller. $d_{50}$=median grain size

<!-- ![](../figures/lecture09/sediment_distribution.png) -->

<img src="../figures/lecture09/sediment_distribution.png" width="800"> 


$log(d_{84})=log(d_{50})+ \sigma \rightarrow d_{84}=d_{50}\exp(\sigma)$

$log(d_{16})=log(d_{50})- \sigma \rightarrow d_{16}=d_{50}/\exp(\sigma)$

If distribution is lognormal, then $\exp(\sigma)=d_{50}/d_{16}=d_{84}/d_{50}$

#### b. Cumulative particle size distribution

Gradation coefficient: $Gr = \frac{1}{2}(\frac{d_{84}}{d_{50}} + \frac{d_{50}}{d_{16}})$

## 2. Particle motion

### 1). Governing equation for a single particle

> $m_s\frac{d\overrightarrow{u}_s}{dt} = \overrightarrow{W} + \overrightarrow{F}_{pressure} + \overrightarrow{F}_{viscous}$

where the different terms are given by:
> Weight: $W = -m_s g\overrightarrow{k}$
>
> Pressure force: $\overrightarrow{F}_{pressure} = - \int_A P \overrightarrow{n} dA$
>
> Viscous force: $\overrightarrow{F}_{viscous} = \int_A T \cdot \overrightarrow{n} dA$
>
where $T$ is the viscous stress tensor, representing viscous stress $\tau$ in all three directions.

### 2). The fall velocity

When a particle falls in still and clear water, it accelerates until it reaches a constant
vertical velocity that is called fall velocity or settling velocity. This velocity can be
assessed from the balance between the downward-directed gravity force $F_G$ (weight minus
buoyancy, which is a pressure force that results from the difference in pressure exerted by a fluid on the particle) and the upward-directed drag force $F_D$.

<!-- ![](../figures/lecture09/fall_velocity.png) -->

<img src="../figures/lecture09/fall_velocity.png" width="400"> 

For a perfect sphere ($V_s = \frac{\pi}{6}d_s^3$; cross-sectional area of the sphere $A_s = \frac{\pi}{4}d_s^2$):

> $F_G = (\rho_s - \rho_0)g(\frac{\pi}{6}d_s^3)$, 
> $F_D = \frac{1}{2}C_D\rho_0 w_s^2 (\frac{\pi}{4}d_s^2)$,

where $C_D$ is the unitless drag coefficient and $w_s$ is the particle fall velocity.

In equilibrium, both forces are in balance and the fall velocity $w_s$  (in m/s) is given by:
> $w_s = \sqrt{\frac{4(s-1)gd_s}{3C_D}}$,

where $s$ is the specific density. 

- A particleâ€™s fall velocity thus depends on **its size, its density and the magnitude of the drag coefficient $C_D$**.

- This drag coefficient depends on the shape of the particle and its roughness, but mainly on the grainâ€™s Reynolds number:
> $Re = w_sd_s/\nu$,

where $\nu$ is the kinematic viscosity coefficient (unit m$^2$/s), which is defined as the dynamic viscosity divided by the water density $\nu=\mu/\rho_0$. A characteristic value for $\nu$ is 10$^{-6}$ m$^2$/s. 
The dynamic viscosity represents
the fluidâ€™s internal resistance to flow (â€˜thicknessâ€™) and is a function of the temperature
and to a smaller extent of the density.

### 3). dependence on Reynolds number

<!-- ![](../figures/lecture09/Cd_Re.png) -->

<img src="../figures/lecture09/Cd_Re.png" width="600"> 

For low grain Reynolds numbers (Re < 0.1 to 0.5) in the so-called **Stokes range**, the drag coefficient can be described by:
> $C_D = 24/Re$

yielding
> $w_s = \frac{(s-1)gd_s^2}{18\nu}$

In this range, the fall velocity depends on **the square of the grain diameter, the relative
density and the kinematic viscosity coefficient**.

For high grain Reynolds numbers (400 < ð‘…ð‘’ < 2 Ã— 105), in the so-called **Newton range**, the drag coefficient becomes a constant ($C_D$ â‰ˆ 0.5). In that case:
> $w_s = 1.6\sqrt{gd_s(s-1)}$

In this range, the fall velocity depends on **the square root of the grain diameter and
the relative density** and is **independent of the kinematic viscosity coefficient**. This is
also the case for extremely high Reynolds numbers (ð‘…ð‘’ > 2 Ã— 105), where the drag
coefficient is (constant) around 0.2.
